```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.cm as cm



# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city=pd.read_csv(city_data_to_load)
ride=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_rides=pd.merge(ride, city, how="left", on="city")
city_rides.head(3)

# Display the data table for preview
city_rides
```

```python
grouped = city_rides.groupby(["type","city"])
```

```python
total_rides=grouped["city"].count()
```

```python
avg_fare=grouped["fare"].mean()
```

```python
df=pd.DataFrame({"total_rides":total_rides, "avg_fare":avg_fare})
```

```python
df2=pd.merge(df, city,how="left", on="city")
```

## Bubble Plot of Ride Sharing Data

```python
fig, ax = plt.subplots()

colors = {'Rural':'gold', 'Suburban':'lightskyblue', 'Urban':'lightcoral'}

grouped = df2.groupby("type")
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='total_rides', y='avg_fare', s=df2['driver_count']*10, 
              alpha=0.75, linewidths=0.5, edgecolor="black", label=key, color=colors[key])
    
ax.set_title("Pyber Rides Sharing Data")
ax.xaxis.set_label_text("Total Number of Rides Per City")
ax.yaxis.set_label_text("Average Fare")
ax.grid()
plt.savefig("bubble_plot.png")
plt.show()
```

## Total Fares by City Type

```python
#Group by type
group_by_type=city_rides.groupby("type")
total_fare=group_by_type["fare"].sum()
```

```python
total_fare_df=city_rides["fare"].sum()
```

```python
# Calculate Type Percents
percentage_of_fares= (total_fare / total_fare_df) * 100
```

```python
plt.title("% of Total Fares by City Type ")
labels=("Rural","Suburban","Urban")
label_colors=('gold',"lightskyblue", "lightcoral" )
explode=(0,0,0.1)
# Build Pie Chart
plt.pie(percentage_of_fares,explode=explode,colors=label_colors, autopct="%1.1f%%", labels=labels, shadow=True,startangle=145)
#plt.axis("equal")
plt.savefig("total_fares_by_city_type.png")
```
